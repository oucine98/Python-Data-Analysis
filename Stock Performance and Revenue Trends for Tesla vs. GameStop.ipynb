```python
import requests
import pandas as pd

# 1. Define the URL and headers
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/115.0.0.0 Safari/537.36"
    )
}

# 2. Fetch the page
response = requests.get(url, headers=headers)
response.raise_for_status()  # will re-raise HTTPError if it failed

# 3. Parse all tables
tables = pd.read_html(response.text)

# 4. Extract and clean the quarterly revenue table
tesla_revenue = tables[0]
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue["Revenue"] = (
    tesla_revenue["Revenue"]
    .str.replace(r"[\$,]", "", regex=True)
    .astype(float)
)

# 5. Display the last five rows
tesla_revenue.tail()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def make_graph(stock_df, revenue_df, title):
    """
    Plots a two-row figure:
      1) Historical share price (line)
      2) Quarterly revenue    (bar)
    """
    # Create 2-row, 1-col subplot with shared x-axis
    fig = make_subplots(
        rows=2, cols=1,
        shared_xaxes=True,
        vertical_spacing=0.2,
        subplot_titles=(
            "Historical Share Price (USD)",
            "Quarterly Revenue (USD Millions)"
        )
    )

    # Row 1: Closing stock price
    fig.add_trace(
        go.Scatter(
            x=stock_df["Date"], 
            y=stock_df["Close"],
            mode="lines",
            name="Close"
        ),
        row=1, col=1
    )

    # Row 2: Revenue bar chart
    fig.add_trace(
        go.Bar(
            x=revenue_df["Date"], 
            y=revenue_df["Revenue"],
            name="Revenue"
        ),
        row=2, col=1
    )

    # Global layout tweaks
    fig.update_layout(
        title=title,
        height=600,
        showlegend=False
    )

    # Y-axis labels
    fig.update_yaxes(title_text="Price (USD)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue (USD Millions)", row=2, col=1)

    fig.show()
```

```python
make_graph(
    stock_df=tesla_data,
    revenue_df=tesla_revenue,
    title="Tesla Stock Price & Quarterly Revenue (2010–2022)"
)
```

```python
make_graph(
    stock_df=gme_data,
    revenue_df=gme_revenue,
    title="GameStop Stock Price & Quarterly Revenue (2010–2022)"
)
```
